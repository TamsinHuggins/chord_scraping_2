```python
# Concat all data files into one df
```

```python
import os
import pandas as pd
```

```python
# Find list of file names
filelist = os.listdir("../scraped_chords_ultimateguitar")
```

```python
# Loop through and concat
df = pd.DataFrame()
for file in filelist:
    newdf = pd.read_csv("../scraped_chords_ultimateguitar/" + file)
    df = pd.concat([df, newdf])
```

```python
df.drop("Unnamed: 0", axis=1)
```

```python
df.to_csv("../all_chords/all_chords.csv")
```
